```python
rdens_tsvd = tsvd_gauss["d_mean"]/(1e12)
rerr_tsvd = np.sqrt(tsvd_gauss["d_var"])/(1e12)

rdens_tik0 = tik0_gauss["d_mean"]/(1e12)
rerr_tik0 = np.sqrt(tik0_gauss["d_var"])/(1e12)

rdens_tik1 = tik1_gauss["d_mean"]/(1e12)
rerr_tik1 = np.sqrt(tik1_gauss["d_var"])/(1e12)

rdens_mf = mf_gauss["d_mean"]/(1e12)
rerr_mf = np.sqrt(mf_gauss["d_var"])/(1e12)
rdens_me = me_gauss["d_mean"]/(1e12)
rerr_me = np.sqrt(me_gauss["d_var"])/(1e12)
error_level = tsvd_gauss["error_level"]

fig,ax = plt.subplots()
fig.set_size_inches((5,3.5))
ax.errorbar(error_level,rdens_tsvd,rerr_tsvd,label = "Truncated SVD",lw=1.5,color="#348ABD")
ax.errorbar(error_level,rdens_tik0,rerr_tik0,label = "0th Order Tikhonov",lw=1.5,color="#A60628")
ax.errorbar(error_level,rdens_tik1,rerr_tik1,label = "1st Order Tikhonov",lw=1.5,color="#7A68A6")
ax.errorbar(error_level,rdens_mf,rerr_mf,label = "Minimum Fisher",lw=1.5,color="#467821")
ax.errorbar(error_level,rdens_me,rerr_me,label = "Maximum Entropy",lw=1.5,color="#D55E00")

ax.tick_params(labelsize="small")
ax.set_xlabel("Error Level",fontsize="large")
ax.set_ylabel("Density Ratio",fontsize="large")
ax.set_ylim(0.7,1.7)
ax.set_xlim(0.0,1.05)
ax.legend(loc=2,frameon=False,fontsize="small",ncol=2)
fig.savefig("gaussian_density_ratio.eps",bbox_inches="tight")
```

```python
def diverging_cmap(cmap,v_min,v_max):
    if isinstance(cmap,str):
        cm = plt.cm.get_cmap(cmap)
    else:
        cm=cmap

    if np.abs(v_max) > np.abs(v_min):
        start = (v_max + v_min)/(2*v_max)
        fin = 1.0
        mid = 1.0 - np.abs(v_min)/(np.abs(v_min)+np.abs(v_max))
    else:
        start = 0.0
        fin = 1.0-(np.abs(v_min) - v_max)/(2*np.abs(v_min))
        mid = np.abs(v_min)/(np.abs(v_min)+np.abs(v_max))
        
    cm = remappedColorMap(cm,start=start,stop=fin,midpoint=mid)
    return cm
```
