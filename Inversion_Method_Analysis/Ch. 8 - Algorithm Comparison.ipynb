```julia
using PyPlot
include("src/netcdf.jl")
```

#Relative Density Plot

##Gaussian

```julia
tsvd = read_ncdf("data/gaussian_tsvd.cdf")
tik0 = read_ncdf("data/gaussian_tik0.cdf")
tik1 = read_ncdf("data/gaussian_tik1.cdf")
mf   = read_ncdf("data/gaussian_mf.cdf")
me   = read_ncdf("data/gaussian_me.cdf")

gaussian = tsvd["dist"]
energy = tsvd["energy"]
pitch = tsvd["pitch"]
npitch = length(pitch)
nenergy = length(energy)
dE = abs(energy[2] - energy[1])
dP = abs(pitch[2] - pitch[1])
A = reshape(sum(tsvd["transfer_matrix"],1),npitch,nenergy)
coverage = A .> eps()
gdens_ratio = (sum(gaussian[coverage])*dE*dP)/(1e12)
```

```julia
error_level = collect(0.05:0.05:1.0)
fig,ax = plt.subplots()
fig[:set_size_inches]((5,3.5))
ax[:errorbar](error_level,tsvd["dens"]/(1e12),tsvd["dens_err"]/(1e12),label = "TSVD",lw=1.5,color="#348ABD")
ax[:errorbar](error_level,tik0["dens"]/(1e12),tik0["dens_err"]/(1e12),label = "0th Order Tikhonov",lw=1.5,color="#A60628")
ax[:errorbar](error_level,tik1["dens"]/(1e12),tik1["dens_err"]/(1e12),label = "1st Order Tikhonov",lw=1.5,color="#7A68A6")
ax[:errorbar](error_level,mf["dens"]/(1e12),mf["dens_err"]/(1e12),label = "Minimum Fisher",lw=1.5,color="#467821")
ax[:errorbar](error_level,me["dens"]/(1e12),me["dens_err"]/(1e12),label = "Maximum Entropy",lw=1.5,color="#D55E00")
ax[:plot](error_level,fill(gdens_ratio,length(error_level)),"k--",lw=1.5,label="Optimal Ratio")
ax[:tick_params](labelsize="small")
ax[:set_xlabel]("Error Level")
ax[:set_ylabel]("Density Ratio")
ax[:set_ylim](0.5,2.1)
ax[:set_xlim](0.0,1.05)
ax[:legend](loc=2,frameon=false,fontsize="small",ncol=2)
```

#Bimax

```julia
tsvd = read_ncdf("data/bimax_tsvd.cdf")
tik0 = read_ncdf("data/bimax_tik0.cdf")
tik1 = read_ncdf("data/bimax_tik1.cdf")
mf   = read_ncdf("data/bimax_mf.cdf")
me   = read_ncdf("data/bimax_me.cdf")

bimax = tsvd["dist"]
energy = tsvd["energy"]
pitch = tsvd["pitch"]
npitch = length(pitch)
nenergy = length(energy)
dE = abs(energy[2] - energy[1])
dP = abs(pitch[2] - pitch[1])
A = reshape(sum(tsvd["transfer_matrix"],1),npitch,nenergy);
coverage = A .> eps();
```
